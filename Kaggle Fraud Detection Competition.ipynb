### 1. Data Discovery

#### 1.1 Load Data and print basic stats

```python
# Train data
train_transaction_df = pd.read_csv('data/train_transaction.csv')
train_identity_df = pd.read_csv('data/train_identity.csv')

# Test data
test_transaction_df = pd.read_csv('data/test_transaction.csv')
test_identity_df = pd.read_csv('data/test_identity.csv')
```

```python
# Take a look at what the transaction data looks like
pd.options.display.max_columns = None
display(train_transaction_df)
```

```python
# Take a look at what the identity data looks like
display(train_identity_df)
```

```python
# Count number of rows and number of columns in each file
train_transaction_row_count, train_transaction_feature_count = train_transaction_df.shape
train_identity_row_count, train_identity_feature_count = train_identity_df.shape
test_transaction_row_count, test_transaction_feature_count = test_transaction_df.shape
test_identity_row_count, test_identity_feature_count = test_identity_df.shape

# Print
print('[train_transaction] Rows: %d, Columns: %d' % (train_transaction_row_count, train_transaction_feature_count))
print('[train_identity] Rows: %d, Columns: %d' % (train_identity_row_count, train_identity_feature_count))
print('\n[test_transaction] Rows: %d, Columns: %d' % (test_transaction_row_count, test_transaction_feature_count))
print('[test_identity] Rows: %d, Columns: %d' % (test_identity_row_count, test_identity_feature_count))
```

```python
# Function to count fraud distribution
def count_fraud(df):
    num_fraud = df[df["isFraud"]==1].shape[0]
    num_fraud_percent = 100. * (float(num_fraud) / float(df.shape[0]))
    return(num_fraud, num_fraud_percent)

# Look at fraud distribution in train set
num_fraud, num_fraud_percent = count_fraud(train_transaction_df)
print('[train_transaction] Num Fraud: %d (%.2f%%)' % (num_fraud, num_fraud_percent))
```

#### 1.2 Form true test and dev sets

```python
# First, create a new df that includes both transaction data and identity data (when available)
joined_df = pd.merge(train_transaction_df, train_identity_df, on='TransactionID', how='outer')
num_rows, num_cols = joined_df.shape
print('Rows: %d, Columns: %d' % (num_rows, num_cols))

# Make sure the new shape makes sense
assert(num_rows == train_transaction_row_count)
assert(num_cols == train_transaction_feature_count + train_identity_feature_count - 1) # Minus 1 bc shared TransactionID
```

```python
# Let's take a look at the new dataframe
display(joined_df)
```

```python
# Choose train set and dev set percentages
train_split = 0.8
dev_split = 1 - train_split

# Use choices above to determine size of sets
train_size = int(num_rows * train_split)
dev_size = num_rows - train_size

# Print
print('Training set size: %d' % train_size)
print('Dev set size: %d' % dev_size)
```

```python
# Randomly choose train and dev sets based on percentage splits
train_set = joined_df.sample(n=train_size)
dev_set = joined_df.loc[~joined_df.index.isin(train_set.index)]

# Make sure sizes make sense
train_rows, train_cols = train_set.shape
dev_rows, dev_cols = dev_set.shape
assert(train_rows == train_size)
assert(dev_rows == dev_size)
```

```python
# Let's make sure that fraudulent transactions are (close to) equally distributed between train and dev sets
num_fraud, num_fraud_percent = count_fraud(train_set)
print('[train_set] Num Fraud: %d (%.2f%%)' % (num_fraud, num_fraud_percent))

num_fraud, num_fraud_percent = count_fraud(dev_set)
print('[dev_set] Num Fraud: %d (%.2f%%)' % (num_fraud, num_fraud_percent))
```

```python
# Split train set and dev set into features (x) and label (y)
train_y = train_set[['TransactionID', 'isFraud']].copy()
train_x = train_set.drop(columns=['isFraud'])

dev_y = dev_set[['TransactionID', 'isFraud']].copy()
dev_x = dev_set.drop(columns=['isFraud'])

# Make sure train sizes make sense
assert(train_x.shape[0] == train_set.shape[0])
assert(dev_x.shape[0] == dev_set.shape[0])
assert(train_x.shape[1] == train_set.shape[1] - 1)
assert(dev_x.shape[1] == dev_set.shape[1] - 1)

# Make sure dev sizes make sense
assert(train_y.shape[0] == train_set.shape[0])
assert(dev_y.shape[0] == dev_set.shape[0])
assert(train_y.shape[1] == 2)
assert(dev_y.shape[1] == 2)
```

```python
# Take a look at train_x (dev_x will look very similar)
display(train_x)
```

```python
# Take a look at train_y (dev_y will look very similar)
display(train_y)
```

```python
# Save new train and dev sets to csv files for easy access later
train_x.to_csv('data/train_x.csv', index=False)
train_y.to_csv('data/train_y.csv', index=False)

dev_x.to_csv('data/dev_x.csv', index=False)
dev_y.to_csv('data/dev_y.csv', index=False)
```
